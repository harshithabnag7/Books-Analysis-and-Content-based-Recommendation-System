## Deleting unwanted columns

```r
books_data <- select(books_data, -volumeInfo.allowAnonLogging)
books_data <- select(books_data, -volumeInfo.contentVersion)
books_data <- select(books_data, -volumeInfo.panelizationSummary.containsEpubBubbles)
books_data <- select(books_data, -volumeInfo.panelizationSummary.containsImageBubbles)
books_data <- select(books_data, -saleInfo.country)
books_data <- select(books_data, -saleInfo.saleability)
books_data <- select(books_data, -saleInfo.isEbook)
books_data <- select(books_data, -accessInfo.country)
books_data <- select(books_data, -accessInfo.viewability)
books_data <- select(books_data, -accessInfo.embeddable)
books_data <- select(books_data, -accessInfo.publicDomain)
books_data <- select(books_data, -accessInfo.textToSpeechPermission)
books_data <- select(books_data, -accessInfo.epub.isAvailable)
books_data <- select(books_data, -accessInfo.accessViewStatus)
books_data <- select(books_data, -accessInfo.quoteSharingAllowed)
books_data <- select(books_data, -searchInfo.textSnippet)
books_data <- select(books_data, -accessInfo.pdf.acsTokenLink)
books_data <- select(books_data, -saleInfo.listPrice.amount)
books_data <- select(books_data, -saleInfo.listPrice.currencyCode)
books_data <- select(books_data, -saleInfo.retailPrice.amount)
books_data <- select(books_data, -saleInfo.retailPrice.currencyCode)
books_data <- select(books_data, -saleInfo.buyLink)
books_data <- select(books_data, -saleInfo.offers)
books_data <- select(books_data, -accessInfo.epub.acsTokenLink)
books_data <- select(books_data, -accessInfo.epub.downloadLink)
books_data <- select(books_data, -accessInfo.pdf.downloadLink)
books_data <- select(books_data, -etag)
books_data <- select(books_data, -selfLink)
books_data <- select(books_data, -kind)
books_data <- select(books_data, -volumeInfo.industryIdentifiers)
books_data <- select(books_data, -volumeInfo.readingModes.text)
books_data <- select(books_data, -volumeInfo.readingModes.image)
books_data <- select(books_data, -volumeInfo.canonicalVolumeLink)
books_data <- select(books_data, -accessInfo.pdf.isAvailable)
books_data <- select(books_data, -accessInfo.webReaderLink)
books_data <- select(books_data, -volumeInfo.printType)
books_data <- select(books_data, -volumeInfo.maturityRating)
```

```r
books_data$book_author <- gsub("\\[|'|\\]", "", books_data$book_author)
```

```r
books_data$book_publisher <- gsub('"', '',  books_data$book_publisher)
```

```r
books_data$book_category <- gsub("\\[|'|\\]", "", books_data$book_category)
```

```r
books_data$book_averageRating <- as.numeric(books_data$book_averageRating)

books_data$book_ratingCount <- as.numeric(books_data$book_ratingCount)

books_data$book_averageRating <- ifelse(is.na(books_data$book_averageRating) | books_data$book_averageRating == "", 0, books_data$book_averageRating)

books_data$book_ratingCount <- ifelse(is.na(books_data$book_ratingCount) | books_data$book_ratingCount == "", 0, books_data$book_ratingCount)
```
